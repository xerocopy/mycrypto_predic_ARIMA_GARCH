#### Creating Bollinger Bands

```python
#df['SMA20'] = df['Close'].rolling(window=20).mean()
# calculate Simple Moving Average with 20 days window
sma = df['Close'].rolling(window=20).mean()

# calculate the standar deviation
rstd = df['Close'].rolling(window=20).std()

upper_band = sma + 2 * rstd
df['upper_band'] = upper_band
lower_band = sma - 2 * rstd
df['lower_band'] = lower_band

#df
```

### Calculate the Simple Moving Average (SMA)


```python
# generated the short medium long longer moving average (2d, 5d, 10d ,50d days respectively)
df['SMA2d'] = df['Close'].rolling(window=2).mean()
df['SMA5d'] = df['Close'].rolling(window=5).mean()
df['SMA10d'] = df['Close'].rolling(window=10).mean()
df['SMA50d'] = df['Close'].rolling(window=50).mean()

df['Signal_LT'] = 0.0
df['Signal_ST'] = 0.0

# replace the nan of the first hundreds of rolls with 0
df.fillna(0.0, inplace =True)

#df.head()
#df[-90:]
#df.dtypes
```

### Generate the trading signal 0 or 1¶


```python
df['Signal_LT'] = np.where(df['SMA2d'].astype('int') > df['SMA5d'].astype('int'), 1.0, 0.0)
df['Signal_ST'] = np.where(df['SMA10d'].astype('int') > df['SMA50d'].astype('int'), 1.0, 0.0)

# Calculate the points in time at which a position should be taken, 1 or -1
df['Entry/Exit'] = df['Signal_LT'].diff()
df.fillna(0.0, inplace =True)
#df.tail()
```

### Generating timeline for reference

```python
date_time = pd.to_datetime(df.index, format='%YYYY,%mm,%dd')
#date_time

timestamp_s = date_time.map(pd.Timestamp.timestamp)

day = 24*60*60
week = 24*60*60*7
month = 24*60*60*7*4.28
quarter = 24*60*60*7*4.28*3
year = 24*60*60*7*4.28*3*4
period = 24*60*60*7*4.28*3*4*4

df['day sin'] = np.sin(timestamp_s * (2 * np.pi / day))
df['day cos'] = np.cos(timestamp_s * (2 * np.pi / day))

df['week sin'] = np.sin(timestamp_s * (2 * np.pi / week))
df['week cos'] = np.cos(timestamp_s * (2 * np.pi / week))

df['month sin'] = np.sin(timestamp_s * (2 * np.pi / month))
df['month cos'] = np.cos(timestamp_s * (2 * np.pi / month))

df['quarter sin'] = np.sin(timestamp_s * (2 * np.pi / quarter))
df['quarter cos'] = np.cos(timestamp_s * (2 * np.pi / quarter))

df['Year sin'] = np.sin(timestamp_s * (2 * np.pi / year))
df['Year cos'] = np.cos(timestamp_s * (2 * np.pi / year))

df['Period sin'] = np.sin(timestamp_s * (2 * np.pi / period))
df['Period cos'] = np.cos(timestamp_s * (2 * np.pi / period))


#df
```

```python
i=1
rate_increase_in_vol=[0]
rate_increase_in_adj_close=[0]

while i<len(df):
    rate_increase_in_vol.append(df.iloc[i]['Volume']-df.iloc[i-1]['Volume'])
    rate_increase_in_adj_close.append(df.iloc[i]['Close']-df.iloc[i-1]['Close'])
    i+=1
    
df['Increase_in_vol']=rate_increase_in_vol
df['Increase_in_adj_close']=rate_increase_in_adj_close
#df
```

#### Add Candlestick information 
(Stock indicators creation)
candle information: 1. Wick length = High - Low;2. Body = abs(Open – Close); 3. Body position = (High+Low)/2 – (open+close)/2; 4. Color = Open – Close >0 red, open – close < 0 green
trend information: 


```python
df['candle_wick'] = df['High']-df['Low']
df['candle_body'] = abs(df['Open']-df['Close'])
df['candle_shape'] = (df['High']+df['Low'])/2-(df['Open']+df['Close'])/2
df['candle_color'] = np.where(df['Open'] > df['Close'],1,0)

#df
```
